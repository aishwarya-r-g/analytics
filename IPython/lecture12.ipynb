```
A = np.array([1,2,3,4])

B = np.array([5,6,7,8])

conditions = np.array([True, True, False, False])
```

```
answer = [A_val if cond else B_val for A_val, B_val, cond in zip(A,B,conditions)]
answer
```

```
answer2 = np.where(conditions,A,B)
answer2
```
